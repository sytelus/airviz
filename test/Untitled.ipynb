```python
import plotly
```

```python
import plotly.graph_objs as go
```

```python
import numpy as np
```

```python
x = np.random.randn(1000)
```

```python
y = np.random.randn(1000)
```

```python
tr = go.Scatter(x=x, y=y, mode='markers')
```

```python
data=[tr]
```

```python
plotly.offline.iplot(data, filename='basic-scatter')
```
